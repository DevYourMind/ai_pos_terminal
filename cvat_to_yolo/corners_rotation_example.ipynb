```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image, ImageOps
import matplotlib.patches as patches
```

```python
xtl=float("710.50") 
ytl=float("1349.94") 
xbr=float("1143.70") 
ybr=float("1434.76") 
rotation=float("358.90")
```

```python
# Load the image
path_image = 'ЗНО0301685576_TS903190613-79401ff5-89c7-42ac-a0d2-4ce926f99379.jpg'
img = ImageOps.exif_transpose(Image.open(path_image))
# Create figure and axis
fig, ax = plt.subplots(figsize=(15, 10))
# Display the image
ax.imshow(img)
# Calculate the center of the bounding box
center_x = (xtl + xbr) / 2
center_y = (ytl + ybr) / 2
# Rotate the rectangle patch around its center
rect = patches.Rectangle(
    (xtl, ytl), xbr - xtl, ybr - ytl, 
    linewidth=2, edgecolor='r', angle=rotation, 
    rotation_point=(center_x, center_y), facecolor='none')
# Add the rectangle patch to the axis
ax.add_patch(rect)
# Show the plot
plt.axis('off')
plt.show()
```

```python
dots = rect.get_corners()
index_x1 = dots[:, 1].argmin()
dot_1 = dots[index_x1]
index_x3 = dots[:, 1].argmax()
dot_3 = dots[index_x3]
index_x2 = dots[:, 0].argmax()
dot_2 = dots[index_x2]
index_x4 = dots[:, 0].argmin()
dot_4 = dots[index_x4]
x1, y1 = dot_1
x2, y2 = dot_2
x3, y3 = dot_3
x4, y4 = dot_4
```

```python
plt.figure(figsize=(15,10))
plt.imshow(img)
plt.plot(*dot_1, 'bo', markersize=1)
plt.plot(*dot_2, 'ro', markersize=1)
plt.plot(*dot_3, 'go', markersize=1)
plt.plot(*dot_4, 'co', markersize=1)
plt.axis('off')
plt.grid()
plt.show()
```
